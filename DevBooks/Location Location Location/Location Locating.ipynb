# Using Geopy to Standardize Location

```python
import pandas as pd
import geopy
```

```python
db=pd.read_excel("Comprehensive TAGS v6.1.9, 2016-Feb 2019.xlsx",sheet_name=1)
```

```python
user_agent="Mozilla/5.0 (Windows NT 5.1) AppleWebKit/535.7 (KHTML, like Gecko) Chrome/16.0.912.63 Safari/535.7"
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="DIL4")
location = geolocator.geocode(db.user_location[1])
location.altitude
location.point
```

```python
d={}
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1.5)
for i in set(db.user_location[db.user_location.notna()]): 
    d[i]=geocode(i)
```

```python
db['user_loc']=db.user_location.map(d)
```

```python
db['address']=db.user_loc[db.user_loc.notna()].map(lambda x: x.address)
```

```python
db['latitude']=db.user_loc[db.user_loc.notna()].map(lambda x: x.latitude)
```

```python
db['longitude']=db.user_loc[db.user_loc.notna()].map(lambda x: x.longitude)
```

```python
db['altitude']=db.user_loc[db.user_loc.notna()].map(lambda x: x.altitude)
```

```python
db['point']=db.user_loc[db.user_loc.notna()].map(lambda x: x.point)
```

```python
db.loc[:,["user_location","address","latitude","longitude","altitude","point"]]
```

```python
db.to_csv("SilentSam Tweets with Std. Location.csv")
```
