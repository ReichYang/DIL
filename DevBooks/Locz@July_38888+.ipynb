```python
db=pd.read_csv("comprehensivestaticforloc_blanksremoved.csv")
```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
out_s= pd.read_excel("out of region To Be Fixed.xlsx", index_col='Unnamed: 0')
```

```python
len(set(out_s['Fixed Location']))
```

```python
d={}
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)
for i in set(out_s['Fixed Location']): 
    d[i]=geocode(i)
```

```python
out_s['cor_user_loc']=out_s['Fixed Location'].map(d)
out_s['cor_address']=out_s.cor_user_loc[out_s.cor_user_loc.notna()].map(lambda x: x.address)
out_s['cor_latitude']=out_s.cor_user_loc[out_s.cor_user_loc.notna()].map(lambda x: x.latitude)
out_s['cor_longitude']=out_s.cor_user_loc[out_s.cor_user_loc.notna()].map(lambda x: x.longitude)
```

```python
out_s.to_excel("corr_out_state.xlsx")
```
