```python
test=loc.drop_duplicates()
```

```python
#address
add_dic=pd.Series(test.address.values,index=test.user_location).to_dict()
```

```python
#lat
lat_dic=pd.Series(test.latitude.values,index=test.user_location).to_dict()
#long
long_dic=pd.Series(test.longitude.values,index=test.user_location).to_dict()
#point
point_dic=pd.Series(test.point.values,index=test.user_location).to_dict()
```

```python
nomloc['addr']=nomloc.no.map(add_dic)
```

```python

nomloc['lat']=nomloc.no.map(lat_dic)
nomloc['long']=nomloc.no.map(long_dic)
nomloc['point']=nomloc.no.map(point_dic)
```
